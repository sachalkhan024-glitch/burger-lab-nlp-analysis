# 🍔 The Burger Lab: Customer Sentiment Analysis

### Executive Summary
* **Objective:** Automate the categorization of unstructured text reviews to determine overall customer satisfaction and identify operational bottlenecks.
* **Methodology:** Built a custom Natural Language Processing (NLP) algorithm in Python to tag raw text reviews as Positive, Negative, or Neutral based on keyword presence.
* **Key Finding:** The restaurant has a critical **40% negative feedback rate**, primarily driven by complaints regarding kitchen QA (cold/undercooked food) and slow wait times. 
* **Recommendation:** Immediate operational intervention in the kitchen is required before leadership focuses on further marketing or expansion.
---

```python
import pandas as pd
```

```python
df=pd.read_csv('burger_lab_reviews.csv')
```

```python
df
```

```python
df['Review_Text']=df['Review_Text'].str.lower()
```

```python
df
```

```python
def sentiment_analyzer(text):
    positive = ['amazing', 'delicious', 'best', 'great', 'good', 'love']
    negative = ['terrible', 'cold', 'undercooked', 'worst', 'awful', 'horrible']
    
    if any(word in text for word in positive):
        return 'Positive'
    elif any (word in text for word in negative):
        return 'Negative'
    else:
        return 'Neutral'
    
```

```python
df['Sentiment']=df['Review_Text'].apply(sentiment_analyzer)
```

```python
df
```

```python
total_sentiment=df['Sentiment'].value_counts()
```
